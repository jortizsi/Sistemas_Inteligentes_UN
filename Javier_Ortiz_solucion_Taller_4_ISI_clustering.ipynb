## 1. Escale los datos
Cree una función que reciba el dataframe con los datos y retorne un arreglo de numpy con los datos. Los datos deben estar escalados de manera que para cada columna el menor valor es 0 y el mayor 1.


```
#Realizado con base en:
#https://towardsdatascience.com/everything-you-need-to-know-about-min-max-normalization-in-python-b79592732b79

from sklearn.preprocessing import MinMaxScaler

def escalar(df):
    '''
    df: dataframe con los datos
    Returns:
    X: arreglo de numpy con los datos escalados
    '''

    ## Your code here

    # Convertir el dataframe a un arreglo multidimensional de numpy
    X = df.values

    # Instanciar un escalador
    scaler = MinMaxScaler()

    # Configurar el escalador segun los valores presentes en el dataset
    scaler.fit(X)

    # generar el nuevo datase normalizado
    X_scaled = scaler.transform(X)

    return X_scaled

```

## 2. Entrene un modelo de agrupamiento

Escriba una función que reciba la matriz de datos y entrene un modelo de clustering. La función debe retornar el modelo entrenado.

```
from sklearn.cluster import KMeans

def clustering(X, k):
    '''
    X: arreglo de numpy con los datos 
    k: número de clusters
    Returns:
    km: instancia de KMeans entrenada con X
    '''
    ## Your code here
    n_clusters = k

    km = KMeans( n_clusters=n_clusters) #instanciar un modelo de clustering con k clusters
    km.fit(X) #entrenar el modelo

    return km
```

```
def num_elems(km, X):
    '''
    km: modelo KMeans preentrenado
    X: arreglo de numpy con los datos 

    Returns:
    num: arreglo de numpy con el número de datos por cluster
    '''
    ## Your code here
    y = km.predict(X) #obtener predicciones
    num = np.array(pd.Series(y).value_counts().sort_index()) #contar los elementos asignados a cada clase
                                                             #Nota: se debe usar sort_index()para ordenar la tabla de frecuencias 
    
    return num

```

Escriba una función que reciba un modelo de KMeans entrenado y retorne un centroide de un cluster particular.

```
def centroid(km, n):
    '''
    km: modelo KMeans preentrenado
    n: índice del cluster

    Returns:
    centroid: arreglo de numpy con el centroide del cluster
    '''
    ## Your code here
    centroid = km.cluster_centers_[n]
    return centroid

```

## 6. Calcule el porcentaje de malos deudores en cada cluster

Escriba una función que calcule el porcentaje de malos deudores en cada uno de los clusters. La función recibe un modelo preentrenado y los datos usados para entrenar el modelo y debe retornar un vector con los porcentajes.

```
def classes(km, X):
    '''
    km: modelo KMeans preentrenado
    X: arreglo de numpy con los datos 

    Returns:
    percentages: arreglo de numpy con el valor de los porcentajes
    '''
    ## Your code here
    numero_de_clusters = km.n_clusters #Obtener cuantos clusters hay usando el atributo del modelo
    matriz_acumuladores =np.zeros((numero_de_clusters, 2)) #Crear una matriz con regisros [buenos_clientes_cluster_i, malos_clientes_cluster_i]
    buenos = 0
    malos  = 1

    percentages = np.zeros(numero_de_clusters)

    es_mal_deudor = X[:,-2] # recortar la columna categorica "class_bad" de malos deudores

    y = km.predict(X) #Obtener las predicciones de a cual cluster se asigna cada persona del dataset
    
    l = len(X) # Numero de personas en el dataset


    #Por cada persona registrada en el dataset determinar su tipo de deudor y su grupo poblacional y actualizar la fila del cluster correspondiente
    for i in range(0,l):
      mal_deudor = es_mal_deudor[i]
      cluster_al_que_pertenece = y[i]
      if (mal_deudor == 1):
        matriz_acumuladores[cluster_al_que_pertenece][malos] += 1
      else:
        matriz_acumuladores[cluster_al_que_pertenece][buenos] += 1


    #Calcular el porcentaje de clientes malos en cada cluster usando la matriz llenada
    for i in range(0,numero_de_clusters):
      total_elementos_cluster_i = matriz_acumuladores[i][malos] + matriz_acumuladores[i][buenos]
      percentages[i] = matriz_acumuladores[i][malos]/total_elementos_cluster_i

    return percentages

```

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans


df = pd.read_csv("credit-german.csv", sep=";")
df = pd.get_dummies(df)
escaler = MinMaxScaler()
X = escaler.fit_transform(df)
km = KMeans(n_clusters=5, random_state=10)
km.fit(X)
result = np.array([0.43589744, 0.95505618, 0.15942029, 0.01265823, 0.16666667])
classes(km, X)
print(np.allclose(classes(km, X), result))
```

# 7. Calcule la característica que más diferencia dos clusters

Escriba una función que calcule la característica que es más diferente entre dos clusters dados. La función debe recibir el dataframe original con los datos, un modelo de clustering preentrenado y los índices de los clusters que se quieren comparar.

```
def most_different(df, km, n, m):
    '''
    df: dataframe con los datos
    km: modelo KMeans preentrenado
    n: índice del primer cluster
    m: índice del segundo cluster
    Returns:
    (feat, val1, val2)
    feat: string con el nombre de la característica
    val1: valor de la característica para el cluster m
    val2: valor de la característica para el cluster n
    '''
    ## Your code here
    centroide_cluster_m = centroid(km,m) #obtener centroide del cluster m
    centroide_cluster_n = centroid(km,n) #obtener centroide del cluster n

    vector_diferencia = abs(centroide_cluster_n - centroide_cluster_m) #calcular vector diferencia de valores de los centroides
    indice_del_valor_maximo = np.argmax(vector_diferencia) #obtener el indice del atributo en que mas se diferencio en los dos clusters
    nombres_columnas_dataframe = df.keys() #armar lista de los nombres de las columnas del dataframe original


    feat = nombres_columnas_dataframe[indice_del_valor_maximo] #nombre de la caracteristica que mas se diferencio
    val1 = centroide_cluster_m[indice_del_valor_maximo] #valor de la caracteristica en el cluster m
    val2 = centroide_cluster_n[indice_del_valor_maximo] #valor de la caracteristica en el cluster n
    return (feat, val1, val2)

```

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans


df = pd.read_csv("credit-german.csv", sep=";")
df = pd.get_dummies(df)
escaler = MinMaxScaler()
X = escaler.fit_transform(df)
km = KMeans(n_clusters=5, random_state=10)
km.fit(X)
res0, res1, res2 = ('property_magnitude_no known property', 0.9743589743589743, 0.043478260869565286)
res = most_different(df, km, 2, 0)
res
print((res0 == res[0]) and np.allclose([res1, res2], res[1:]))
```

# 8. Calcule el número de clusters usando el método del codo

Diseñe una función que reciba un modelo de kmeans (no preentrenado) y encuentre un valor *óptimo* para el número de clusters usando la estrategia del codo. El valor óptimo es el máximo valor de $k$ tal que el incremento de la inercia entre $k-1$ y $k$ es al menos de un 2%. Su función no debe crear una nueva instancia del modelo de kmeans. Debe usar la función `km.set_params(n_clusters=k)` para fijar el número de clusters.

```
def elbow(km, X, n):
    '''
    km: modelo KMeans
    X: arreglo de numpy con los datos
    n: valor máximo del número de clusters a explorar

    Returns:
    (k, inertia)
    k: número óptimo de clusters
    inertia: valor de la inercia correspondiente a k clusters
    '''
    ## Your code here

    
    K = range(3,n+1) # obtener lista con los posibles k's a probar
    k = 0            # variable para guardar la solucion numero de clusters optimo

    km.set_params(n_clusters=2) # el caso inicial es un modelo con dos clusters
    km = km.fit(X)              # entrenar el modelo
    inercia_k_menos_uno_clusters = km.inertia_ # variable para la inercia con k-1
    inercia_k_clusters = 0 # variable para la inercia con k
    inertia = 0 # variable para guardar la solucion de inercia segun el numero de clusters optimo



    for i in K:

        km.set_params(n_clusters=i) # establecer el numero de clusters
        km = km.fit(X) #entrenar el modelo
        inercia_k_clusters = km.inertia_ # obtener la inercia del modelo acual

        cambio = 1.0 - (inercia_k_clusters / inercia_k_menos_uno_clusters) #calcular mejora porcentual lograda con el paso actual
        
        if not (cambio >= 0.02): # cuando no se cumpla la condicion definida salir del ciclo for
          break
        
        inercia_k_menos_uno_clusters = inercia_k_clusters #la inercia k de esta iteracion sera la k-1 de la siguiente
        k = i # guardar el ultimo valor k. si en la siguiente iteracion la condicion de salida se cumple, quedara este resultado pues no se actualizara k de nuevo



    inertia = inercia_k_menos_uno_clusters # el ultimo k hizo la condicion desalida verdadera entonces retornamos la inercia k-1

    return (k, inertia)
```

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans


df = pd.read_csv("credit-german.csv", sep=";")
df = pd.get_dummies(df)
escaler = MinMaxScaler()
X = escaler.fit_transform(df)
km = KMeans(random_state=10)
res = np.array((14, 501.65992863324675))
elbow(km, X[:100], 20)
print(np.allclose(res, elbow(km, X[:100], 20)))
```
